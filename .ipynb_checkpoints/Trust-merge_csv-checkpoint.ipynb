```python
import os 
import pandas as pd
main_dir = os.getcwd()
print(main_dir)
```

```python
# Replace with the actual path to your folder
folder_path = main_dir+"/final_data/"  

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

# Create an empty list to store individual DataFrames
dataframes = []

# Read each CSV file and append its DataFrame to the list
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
initial_df = pd.concat(dataframes)

# Reset the index of the concatenated DataFrame
initial_df.reset_index(drop=True, inplace=True)

# Print the resulting DataFrame
#print(concatenated_df)
```

```python
print(len(initial_df))
```

```python
#save initial dataframe with all the users and data
initial_df.to_csv('all_fixations.csv', index=False)
```

```python
#chekc the number of participants in the initial dataframe
grouped_df = initial_df.groupby('participant_id')
print(len(grouped_df))
```

```python
final_df = pd.DataFrame()
```

```python
final_df = pd.DataFrame(columns=['participant_id', 'task_number', 'bug_number',
                           'fixation_rate_AOI_A1','fixation_rate_AOI_A2','fixation_rate_AOI_A3',
                           'fixation_rate_AOI_A4','fixation_rate_AOI_A5','fixation_rate_AOI_A6',
                           'avg_fixation_time_AOI_A1','avg_fixation_time_AOI_A2','avg_fixation_time_AOI_A3',
                           'avg_fixation_time_AOI_A4','avg_fixation_time_AOI_A5','avg_fixation_time_AOI_A6',
                           'is_accepted','is_correct'])
```

```python
for index, row in initial_df.iterrows():
    
    user_id = row['participant_id']
    task_number = row['task_number']
    aoi = row['AOI']
    if(row['AOI'] != 'All'):
        condition = (final_df['participant_id'] == user_id) & (final_df['task_number'] == task_number)

        #create a new row
        if final_df.loc[condition].empty:

            new_row = {'participant_id': user_id, 'task_number': task_number, 'bug_number': row['bug_number'],
                      'is_accepted':row['is_accepted'],'is_correct':row['is_correct']}
            new_row['fixation_rate_AOI_' + aoi] = row['fixation_rate']
            new_row['avg_fixation_time_AOI_' + aoi] = row['avg_fixation_time']
            final_df = final_df.append(new_row, ignore_index=True)
            #final_df = pd.concat([df, new_row], ignore_index=True)


        #update the existing row
        else:
            final_df.loc[condition, 'fixation_rate_AOI_' + aoi] = row['fixation_rate']
            final_df.loc[condition, 'avg_fixation_time_AOI_' + aoi] = row['avg_fixation_time']

 
```
